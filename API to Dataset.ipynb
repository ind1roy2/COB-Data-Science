```python
# importing the necessary libraries
import requests
import pandas as pd
```

```python
# fetching the url using requests
url = "https://data.binance.com/api/v3/ticker/24hr"
r = requests.get(url)
```

```python
# validating the request
if r.status_code == 200:
    dataset = r.json()
```

```python
# Converting to pandas dataframe
df = pd.DataFrame(dataset)
```

```python
# Saving the file
df.to_csv("api_to_dataset.csv",index = False)
```
